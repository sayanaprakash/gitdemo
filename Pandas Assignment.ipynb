### 4. Find the index positions of items common in  series A and series B using numpy.where() and pd.Index() methods?

```python
A = pd.Series([1,2,3,4,5,6,7,8])
B = pd.Series([1,2,3,7,9])

# Using np.where() method

result = []
for i in B:
    if i in A:
        result.append(int(np.where(i==A)[0]))
print(result)
```

```python
# Using pd.Index() method

result = []
for i in B:
    if i in A:
        result.append(pd.Index(A).get_loc(i))
print(result)
```

### 5. How to convert the first character of each Categorical value in a series to uppercase keeping rest of values to lowercase?

```python
series5 = pd.Series(['thrissur', 'trivandrum', 'kollam', 'alappuzha', 'kannur', 'wayanad'])
series5 = series5.map(lambda x: x[0].upper() + x[1:])
series5
```

### 7. Analysing earthquake data using url: https://raw.githubusercontent.com/ammishra08/MachineLearning/master/Datasets/earthquakes_2014.csv

##### a. Create a DataFrame using URL

```python
earthquake_df = pd.read_csv("https://raw.githubusercontent.com/ammishra08/MachineLearning/master/Datasets/earthquakes_2014.csv", 
                           sep=',', header = 0)
earthquake_df
```

##### b. Display Info of the Dataframe

```python
earthquake_df.info()
```

##### c. Convert time column from object to datetime format 
    [Hint: Use pd.to_datetime()]

```python
earthquake_df.time = pd.to_datetime(earthquake_df.time)
earthquake_df.info()
```

##### d. Rename columns latitude, longitude, depth, mag to  ‘Lat’, ‘Long’, ‘D, M’ column names.

```python
earthquake_df = earthquake_df.rename(columns={'latitude':'Lat', 'longitude':'Long', 'depth':'D', 'mag':'M'})
earthquake_df.info()
```

##### e. Find out invalid data (the nan values from missing points)

```python
earthquake_df.isnull()
```

##### f. Drop all nan values from the existing data.

```python
earthquake_df.dropna()
```

##### g. Replace the invalid data with central tendencies.

```python
# earthquake_df['gap'].fillna(value=earthquake_df['gap'].mean(), inplace=True)
# earthquake_df
earthquake_df[['gap','dmin','rms']]=earthquake_df[['gap','dmin','rms']].fillna(value=earthquake_df[['gap','dmin','rms']].mean())
earthquake_df[['D','M','nst']]=earthquake_df[['D','M','nst']].fillna(value=earthquake_df[['D','M','nst']].median())
earthquake_df[['magType']]=earthquake_df[['magType']].fillna(value=earthquake_df[['magType']].mode())
earthquake_df
```
